```python
import pandas as pd

df = pd.read_csv('claim_data.csv')

approval_rate = round(
    (df['Claim Status'] == 'Paid').sum() / len(df) * 100, 2
)
print(f"KPI 1 | Approval Rate: {approval_rate}%")

print("KPI 2 | Avg Days to Decision: ⚠ Blocked — needs Decision Date from FS-06")

fraud_codes = ['Duplicate claim', 'Incorrect billing information']
fraud_rate = round(
    df['Reason Code'].isin(fraud_codes).sum() / len(df) * 100, 2
)
print(f"KPI 3 | Fraud Flag Rate: {fraud_rate}%")

auto = df[(df['Outcome'] == 'Paid') & (df['Claim Status'] == 'Paid')]
auto_decision_rate = round(len(auto) / len(df) * 100, 2)
print(f"KPI 4 | Auto-Decision Rate (proxy): {auto_decision_rate}%")
print("       ⚠ Replace with ML-04 auto_decision_flag when available")

df['amount_saved'] = df['Billed Amount'] - df['Paid Amount']
total_saved = df['amount_saved'].sum()
avg_saved_per_claim = round(df['amount_saved'].mean(), 2)
print(f"KPI 5 | Total £ Saved: £{total_saved:,}")
print(f"       Avg per Claim: £{avg_saved_per_claim}")

print("\n── KPI SUMMARY ──────────────────────────────────────")
print(f"  Approval Rate       : {approval_rate}%")
print(f"  Avg Days to Decision: BLOCKED (no Decision Date)")
print(f"  Fraud Flag Rate     : {fraud_rate}%")
print(f"  Auto-Decision Rate  : {auto_decision_rate}% (proxy)")
print(f"  £ Saved             : £{total_saved:,}")
```
